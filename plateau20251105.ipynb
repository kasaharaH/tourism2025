```python
import ipywidgets as widgets
from plateaukit.core.widgets.interactive_deck import InteractiveDeck


def show_widget():
      slider = widgets.FloatSlider(
                    value=0.0,
                    min=0,
                    max=12.0,
                    step=0.01,
                    description="0.5m浸水後経過時間（h）",
                    layout=widgets.Layout(width="100%"),
      )

      output = widgets.Output()
      deck = InteractiveDeck(area.gdf)
      with output:
            display(deck.deck)

      @output.capture()
      def handler(slider_value):
            # ①
            gdf = area.gdf.copy()
            # ②
            elapsed_time = slider_value
            in_duration = elapsed_time <= gdf["riverFloodingRiskDuration"]
            # ③
            gdf["fill_color"] = in_duration.map({True: [255, 0, 255], False: None})
             # ④
            updated_layer = deck._create_building_layer(gdf)
            deck.deck.layers = [updated_layer]
            deck.deck.update()

      widgets.interactive(handler, slider_value=slider)
      display(slider, output)

show_widget()
```

```python
# --- 1) 前処理：riverFloodingRisk を正規化して列に展開 -----------------------
import pandas as pd
import numpy as np
import json

def to_record(x):
    """riverFloodingRisk の各行を dict に揃える（NaN/文字列/リスト/一要素辞書を許容）"""
    if pd.isna(x):
        return None
    if isinstance(x, list):
        if len(x) == 0:
            return None
        x = x[0]
    if isinstance(x, str):
        try:
            x = json.loads(x)
        except Exception:
            return None
    if isinstance(x, dict):
        # {"流域名": {...}} の形なら中身を取り出し、流域名は 'basin' 列で保持
        if len(x) == 1 and isinstance(next(iter(x.values())), dict):
            basin, rec = next(iter(x.items()))
            rec = rec.copy()
            rec["basin"] = basin
            return rec
        return x
    return None

s = area.gdf["riverFloodingRisk"]
records = s.map(to_record)

risk = pd.json_normalize(
    [r if isinstance(r, dict) else {} for r in records],
    sep="_"
)
risk.index = area.gdf.index

# キー名のゆらぎに対応（depth/floodingDepth, duration/floodingDuration）
def coalesce(cols):
    exist = [c for c in cols if c in risk.columns]
    return risk[exist[0]] if exist else pd.Series(np.nan, index=risk.index)

depth    = coalesce(["depth", "floodingDepth"])
duration = coalesce(["duration", "floodingDuration"])
scale    = coalesce(["scale", "floodingRiskScale", "riverFloodingRisk_scale"])

area.gdf["flood_depth_m"]   = pd.to_numeric(depth, errors="coerce")
area.gdf["flood_duration_h"] = pd.to_numeric(duration, errors="coerce")
area.gdf["flood_scale"]     = scale

# 可視対象（深さ>0 & 継続時間>0）のみ
gdf = area.gdf.copy()
gdf_valid = gdf.query("flood_depth_m > 0 and flood_duration_h > 0").copy()
if gdf_valid.empty:
    raise ValueError("可視化対象が空です。対象エリアやsize、scale条件を見直してください。")
```

```python
# --- 2) アニメーション本体：t 時間における浸水継続中の建物を描く ------------
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 時間グリッド（0時間〜最大継続時間、整数時間で）
t_max = int(np.nanmax(gdf_valid["flood_duration_h"]))
times = np.arange(0, max(1, t_max + 1))  # 0,1,2,...,t_max

# ベース図（全建物の外形は薄く、浸水中は濃く）
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_aspect('equal')
ax.set_title("Flooding Duration Animation", fontsize=12)

# 背景：全建物（薄い色）
gdf.boundary.plot(ax=ax, color="#bbbbbb", linewidth=0.3)

# 初期フレーム：t=0（継続>0の建物を塗る）
flooded = gdf_valid.copy()
flooded_now = flooded[flooded["flood_duration_h"] >= 0]
# プロットハンドルを保持して、フレームごとに描き直す
coll = flooded_now.plot(ax=ax, color="#1f77b4", linewidth=0, alpha=0.6)

# 累積カウンタ用テキスト
txt = ax.text(0.01, 0.99, "", transform=ax.transAxes, va="top", ha="left", fontsize=11)

def update(frame_t):
    global coll
    # 前フレームを消去して再描画
    for c in (coll.collections if hasattr(coll, "collections") else []):
        c.remove()
    # t 時間以上浸水が続く建物（=まだ水に浸かっている）
    flooded_now = flooded[flooded["flood_duration_h"] >= frame_t]
    coll = flooded_now.plot(ax=ax, color="#1f77b4", linewidth=0, alpha=0.6)
    txt.set_text(f"t = {frame_t} h  |  flooded buildings: {len(flooded_now):,}")
    return coll, txt

anim = FuncAnimation(fig, update, frames=times, interval=600, blit=False, repeat=True)

# Notebook で再生（HTML5）
HTML(anim.to_jshtml())
```
